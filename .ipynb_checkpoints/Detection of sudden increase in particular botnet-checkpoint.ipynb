We assume that the ability to determine if an IP belongs to a particular botnet has been predetermined and the problem focuses on the detection of a sudden increase in such a botnet.

A real life example of such a trend can be seen from the WireX Android Botnet. News of WireXâ€™s emergence first surfaced August 2, 2017, when a modest collection of hacked Android devices was first spotted conducting some fairly small online attacks. Less than two weeks later, however, the number of infected Android devices enslaved by WireX had ballooned to the tens of thousands. We can see a change in the time series at around 7 Aug 2017.

*Source:https://krebsonsecurity.com/2017/08/tech-firms-team-up-to-take-down-wirex-android-ddos-botnet/*

In time series analysis, such a change can be detected using changepoint detection. A change point is a location where there is a change in the distribution in a given time series. It is a related by slightly different problem from anomaly detection, where the purpose is to identify observations that do not conform to an expected pattern or distribution in the data. In change point detection, we assume that the data before the change point conforms to one distribution, while data after the change point comes from a second, different distribution and we are interested in identifying where the change has occured, as soon as possible after it occurs.

To demonstrate how this can be applied, we leverage a dataset provided by Netlab360 that contains a list of Mirai IPs from 1 Aug 2016 to 8 Aug 2016. *http://data.netlab.360.com/mirai-scanner/* 

We leverage on a change point detection algorithm developed at Twitter - E-Divisive with Medians (EDM), which employs energy statistics to detect divergence in median. *https://blog.twitter.com/engineering/en_us/a/2014/breakout-detection-in-the-wild.html*   *Code:https://github.com/roland-hochmuth/BreakoutDetection*

```python
import pandas as pd
import breakout_detection
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("fivethirtyeight")
```

```python
# Load data and perform some cleaning
filepath = "scanner.list"
scanner_df = pd.read_csv(filepath, sep="\t", header=None)
scanner_df.columns = ["timestamp", "sip", "dport"]
scanner_df.sip = scanner_df.sip.apply(lambda x: x.split('=')[1])
scanner_df.dport = scanner_df.dport.apply(lambda x: x.split('=')[1])
scanner_df.timestamp = pd.to_datetime(scanner_df.timestamp)
scanner_df = scanner_df.set_index('timestamp')
scanner_df.head()
```

```python
# Visualise the number of unique IPs bin by the hour
plot_df = scanner_df.sip.resample('H').nunique()
ax = plot_df.plot(figsize=(15,5), title="Unique Number of Mirai IPs", fontsize=12, x_compat=True)
ax.set_ylabel('Count', fontsize=12)
ax.set_xlabel('Datetime', fontsize=12)
```

Visually, we can see 2 points in the graph where there is a sudden increase in the number of unique IPs around 2 Aug and 6 Aug. Let's see if EDM is able to detect the change points. There are a few parameters that can be tuned.

- **min_size**: The minimum number of observations between change points. This is essentially the amount of observations that you want to use as your baseline.
- **degree**: The degree of the penalization polynomial. Can take the values 0, 1 or 2.
- **beta**: Used to further control the amount of penalization. The larger the value the larger the penalisation the larger the change has to be before it can be considered a change point.

```python
# scale data between 0 and 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Z = scaler.fit_transform(plot_df.to_frame())
```

```python
# find level shifts - taking 24 hours as a baseline
edm_multi = breakout_detection.EdmMulti()
edm_multi.evaluate(Z.flatten(), min_size=24, beta=0.01, degree=1)
```

```python
# Visualise
plot_df = scanner_df.sip.resample('H').nunique()
ax = plot_df.plot(figsize=(15,5), title="Unique Number of Mirai IPs", fontsize=12, x_compat=True)
for i in edm_multi.getLoc(): 
    plt.axvline(plot_df.index[i], color='k')
ax.set_ylabel('Count', fontsize=12)
ax.set_xlabel('Datetime', fontsize=12)
```

Using the past 24 hours as the baseline, EDM was able to detect 3 change points, the first that corresponded to the initial increase early on. The second point actually corresponded to a drop in the number of IPs, followed by the last change point where the trend starts to increase again. This shows that EDM can be used to not only detect for sudden increase but sudden drops as well to signal that a threat has possibly subsided. We can fine-tune this further if we do not want so many alerts and perhaps only consider the point around 6 Aug to be a sudden increase by modifying both the beta and min_size values.

```python
# find level shifts - taking 48 hours as a baseline
edm_multi = breakout_detection.EdmMulti()
edm_multi.evaluate(Z.flatten(), min_size=48, beta=0.009, degree=1)
```

```python
# Visualise
plot_df = scanner_df.sip.resample('H').nunique()
ax = plot_df.plot(figsize=(15,5), title="Unique Number of Mirai IPs", fontsize=12, x_compat=True)
for i in edm_multi.getLoc(): 
    plt.axvline(plot_df.index[i], color='k')
ax.set_ylabel('Count', fontsize=12)
ax.set_xlabel('Datetime', fontsize=12)
```

We now see that EDM has only detected a single change point on 6 Aug 2016.

Based on past botnet reports, we infection usually origin from a particular ISP from a particular country and then starts to propergate. As such, we can consider to perform this modelling on a per country basis, which would increase the fidelity of the signals and accuracy of the detection. 

In this analysis we have used unique IPs as a metric for measuring botnet size. However, it may not be the best metric due to DHCP churn [1]. To account for this, a better metric maybe the number of hosts actively "scanning" at the start of every hour. In [2], IPs were considered active if they were scanning at a rate of at least five packets per second, expiring inactive scans after 20 minutes. Another metric that can provide an indication of the relative size of the botnet is the amount of C2 DNS lookups.
